```python
from openai import OpenAI
import requests
from bs4 import BeautifulSoup
```

```python
#basta colocar a url aqui e dar play
url = 'https://huyenchip.com/'
```

```python
def fetch_website_contents(url):
    """
    retorna títulos e conteúdos do website;
    limite de 2 mil caracteres de conteúdo
    """
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.title.string if soup.title else "Título não encontrado"
    if soup.body:
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
    return (title + text)[:2_000]
```

```python
mensagem = [
    {'role':'system','content':'Faça um resumo dessa página e remova os \\n para um texto mais bonito'},
    {'role':'user','content':fetch_website_contents(url)}
]
display(mensagem)
```

```python
ollama_base_url = "http://localhost:11434/v1"

ollama = OpenAI(api_key="ollama", base_url= ollama_base_url)

response = ollama.chat.completions.create(
    model='llama3.2',
    messages=mensagem
)

response.choices[0].message.content
```
