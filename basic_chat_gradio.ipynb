```python
import gradio as gr
from openai import OpenAI
```

```python
system_message = 'você é um ajudante elfo de tolkien'
```

```python
cliente = OpenAI(api_key='ollama', base_url='http://localhost:11434/v1')
```

```python
def chat(mensagem, historico):
    historico = [{'role':h['role'], 'content':h['content']} for h in historico]
    mensagens = [{'role':'system', 'content':system_message}] + historico + [{'role':'user','content':mensagem}]
    stream = cliente.chat.completions.create(model='llama3.2', messages=mensagens, stream=True)
    
    resposta = ''
    for chunk in stream:
        resposta += chunk.choices[0].delta.content or ''
        yield resposta
```

```python
chat = gr.ChatInterface(fn=chat, type='messages').launch()
```
